```python
def generate_random_img(low,high):
    random_img_array = np.random.randint(low,high,size=(299,299,3))
    random_img = image.array_to_img(random_img_array,scale=False)
    return random_img

def generate_random_img_colors(lows,highs):
    channel_1 = np.random.randint(lows[0],highs[0],size=(299,299,1))
    channel_2 = np.random.randint(lows[1],highs[1],size=(299,299,1))
    channel_3 = np.random.randint(lows[2],highs[2],size=(299,299,1))
    img_array = np.concatenate((channel_1,channel_2,channel_3),axis=2)
    random_img = image.array_to_img(img_array,scale=False)
    return random_img
```
